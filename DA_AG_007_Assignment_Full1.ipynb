## Question 1
What is hypothesis testing in statistics?

### Answer:
Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It allows us to test assumptions (called hypotheses) about population parameters. The process involves stating a null hypothesis (H₀) and an alternative hypothesis (H₁), collecting data, and then using statistical tests to determine whether to reject H₀.

## Question 2
What is the null hypothesis, and how does it differ from the alternative hypothesis?

### Answer:
The null hypothesis (H₀) assumes there is no effect or no difference in the population, while the alternative hypothesis (H₁) suggests that there is an effect or difference. Hypothesis testing evaluates whether the sample data provides enough evidence to reject H₀ in favor of H₁.

## Question 3
Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

### Answer:
The significance level (α) is the probability of making a Type I error, i.e., rejecting the null hypothesis when it is actually true. It is usually set at 0.05 (5%). If the calculated p-value is less than α, we reject the null hypothesis. Thus, α serves as the threshold for making statistical decisions.

## Question 4
What are Type I and Type II errors? Give examples of each.

### Answer:
- **Type I Error**: Rejecting a true null hypothesis (false positive). Example: concluding a medicine works when it does not.
- **Type II Error**: Failing to reject a false null hypothesis (false negative). Example: concluding a medicine does not work when it actually does.

## Question 5
What is the difference between a Z-test and a T-test? Explain when to use each.

### Answer:
- A **Z-test** is used when the population variance is known and the sample size is large (n > 30).
- A **T-test** is used when the population variance is unknown and/or the sample size is small (n ≤ 30).
Both tests help compare sample means with population means, but they apply under different conditions.

## Question 6
Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
# Answer
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

## Question 7  
Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.  

Dataset:  
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,  
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,  
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,  
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]  


```python
# Answer
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z statistic
z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)
```

## Question 8  
Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.  


```python
# Answer
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Mean and std
mean = np.mean(data)
std_err = stats.sem(data)

# 95% confidence interval
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("Mean:", mean)
print("95% Confidence Interval:", conf_int)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Question 9  
Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.  


```python
# Answer
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate sample data
data = np.random.normal(100, 15, 1000)

# Calculate Z-scores
z_scores = zscore(data)

# Print first 10 z-scores
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
